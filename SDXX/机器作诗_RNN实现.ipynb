```python
import pandas as pd
```

### 文本数据处理

```python
poems_text = pd.read_table("poems_clean.txt",header=None) # 读取文本数据
poems_text.columns=["text"]  # 设置列名称
poems_text.head()
```

```python
poems = []
for line in poems_text["text"]:  # 遍历每一首诗
    title,poem = line.split(":")  # 将每首诗歌拆分成标题与诗歌内容
    poem = poem.replace(" ","")  # 将空格替换为空字符串
    poem = poem.replace('\n','')  # 将换行符也去掉
    poems.append(list(poem))
```

```python
poems[0]
```

```python
from keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer()
tokenizer.fit_on_texts(poems)  # 给每个不重复的字编索引
poems_digit = tokenizer.texts_to_sequences(poems)
print(tokenizer.word_index)

# Tokenizer默认把0索引留给停用词了，其他的词是从索引1开始的
vacab_size = len(tokenizer.word_index) + 1
```

```python
poems_digit
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences



poems_digit = pad_sequences(poems_digit,maxlen=50,padding="post")
print("原始诗歌")
print(poems[3864])
print("\n")
print("编码+补全后的结果")
print(poems_digit[3864])
```

```python
X = poems_digit[:,:-1]  # 特征
Y = poems_digit[:,1:]   # 标签

print(poems_digit.shape)
print(X.shape)
print(Y.shape)
```

```python
print("X示例","\t","Y示例")

for i in range(10):
    print(X[0][i],"\t",Y[0][i])
```

```python
vacab_size
```

```python
from keras.utils import to_categorical
# 对标签进行one-hot编码
Y = to_categorical(Y,num_classes=vacab_size,dtype='int8')
Y
```

```python
Y.shape
```

## 模型设计

```python
from keras.models import Model
from keras.layers import Input,Dense,Embedding,Activation,SimpleRNN
```

```python
embedding_size = 64   # 词向量的维度数
hidden_size = 128   # 隐藏层的神经元数量

inp = Input(shape=(49,))
x = Embedding(vacab_size,embedding_size,mask_zero=True)(inp)
x = SimpleRNN(hidden_size,return_sequences=True)(x)

x = Dense(vacab_size)(x)
pred = Activation('softmax')(x)

model = Model(inp,pred)
model.summary()
```

```python
from keras.optimizers import Adam


model.compile(loss='categorical_crossentropy',optimizer=Adam(lr=0.01),metrics=['accuracy'])
model.fit(X,Y,epochs=10,batch_size=128,validation_split=0.2)
```
